```python
# find total sales each month from the given unit sold matrix and price per unit matrix
```

```python
unit_sold = [[50, 60, 25], [10, 13, 5], [40, 72, 52]]
price_per_unit = [20, 30, 15]
```

```python
import numpy as np
```

```python
total_sales = np.dot(unit_sold, price_per_unit)
total_sales
```
